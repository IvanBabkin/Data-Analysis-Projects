```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Covid Data 
#### (From European Centre for Disease Prevention and Control)

```python
covid_dataset = pd.read_excel('COVID-19-geographic-disbtribution-worldwide.xlsx')
covid_dataset
```

```python
drop_columns = ['day','month','year','geoId','countryterritoryCode','popData2019','continentExp','Cumulative_number_for_14_days_of_COVID-19_cases_per_100000']
covid_dataset.drop(drop_columns, axis = 1, inplace = True)
covid_dataset.rename(columns={'dateRep':'Date', 'countriesAndTerritories':'Countries'}, inplace=True)
covid_dataset
```

```python
covid_data = covid_dataset.pivot(index='Countries', columns='Date', values=['cases','deaths'])
covid_data
```

```python
covid_data.columns
```

```python
countries = list(covid_data.index)
countries_new = list()
for c in countries:
    if c.find('_') == -1:
        countries_new.append(c)
        continue
    country = ' '.join(c.split('_'))
    countries_new.append(country)
```

```python
len(countries) == len(countries_new)
```

```python
covid_data['Countries'] = countries_new
covid_data.set_index('Countries', inplace = True)
covid_data
```

## Daily Cases

```python
covid_data_cases = covid_data.drop('deaths', axis = 1)
covid_data_cases
```

```python
covid_data_cases.shape
```

### Daily Cases Visualisation

```python
covid_data_cases.loc["China"].plot(figsize = (15,8))
covid_data_cases.loc["United States of America"].plot(figsize = (15,8))
covid_data_cases.loc["Russia"].plot(figsize = (15,8))
plt.legend();
```

### Change in daily cases

```python
covid_data_cases.loc['China'].diff().plot(figsize = (15,8))
covid_data_cases.loc['United States of America'].diff().plot(figsize = (15,8))
covid_data_cases.loc['Russia'].diff().plot(figsize = (15,8))
plt.legend();
```

```python
print('Average cases per day in China:',round(covid_data_cases.loc['China'].mean(),3))
print('Average cases per day in USA:',round(covid_data_cases.loc['United States of America'].mean(),3))
print('Average cases per day in Russia:',round(covid_data_cases.loc['Russia'].mean(),3))
```

## Daily Deaths

```python
covid_data_deaths = covid_data.drop('cases', axis = 1)
covid_data_deaths
```

### Daily Deaths Visualisation

```python
covid_data_deaths.loc["China"].plot(figsize = (15,8))
covid_data_deaths.loc["United States of America"].plot(figsize = (15,8))
covid_data_deaths.loc["Russia"].plot(figsize = (15,8))
plt.legend();
```

### Change in daily deaths

```python
covid_data_deaths.loc['China'].diff().plot(figsize = (15,8))
covid_data_deaths.loc['United States of America'].diff().plot(figsize = (15,8))
covid_data_deaths.loc['Russia'].diff().plot(figsize = (15,8))
plt.legend()
```

```python
print('Average deaths per day in China:',round(covid_data_deaths.loc['China'].mean(),3))
print('Average deaths per day in USA:',round(covid_data_deaths.loc['United States of America'].mean(),3))
print('Average deaths per day in Russia:',round(covid_data_deaths.loc['Russia'].mean(),3))
```

## Average daily cases and deaths

```python
data_index = list(covid_data_cases.index)
data_index
```

```python
avg_cases = list()
avg_deaths = list()
for country in data_index:
    avg_cases.append(covid_data_cases.loc[country].mean())
    avg_deaths.append(covid_data_deaths.loc[country].mean())
covid_data_cases['Average Cases Per Day'] = avg_cases
covid_data_deaths['Average Deaths Per Day'] = avg_deaths
```

```python
covid_data_cases.head()
```

```python
covid_data_deaths.head()
```

```python
covid_average_data = pd.concat([covid_data_cases['Average Cases Per Day'], covid_data_deaths['Average Deaths Per Day']], axis=1)
covid_average_data
```

```python
covid_average_data.corr()
```

```python
cases = covid_average_data['Average Cases Per Day']
deaths = covid_average_data['Average Deaths Per Day']
graph1 = sns.scatterplot(cases, deaths)
graph1.figure.set_size_inches(18.5, 10.5)
sns.despine()
```

```python
graph2 = sns.regplot(np.log(cases), deaths)
graph2.figure.set_size_inches(18.5, 10.5)
sns.despine()
```

```python
graph3 = sns.regplot(np.log(cases), np.log(deaths))
graph3.figure.set_size_inches(18.5, 10.5)
sns.despine()
```

### Comparing to other varibales

```python
hdi_data = pd.read_csv('Human_development_index_(HDI).csv')
hdi_data
```

```python
hdi = pd.concat([hdi_data['Country'], hdi_data['2018']], axis=1)
hdi.rename(columns={'2018':'HDI'}, inplace=True)
pd.to_numeric(hdi["HDI"], downcast="float")
hdi.set_index('Country', inplace = True)
hdi
```

```python
hdi_covid = covid_average_data.join(hdi, how="inner")
hdi_covid
```

```python
cases2 = hdi_covid['Average Cases Per Day']
deaths2 = hdi_covid['Average Deaths Per Day']
HDI_v = hdi_covid['HDI']
```

```python
graph4 = sns.regplot(HDI_v, np.log(cases2))
graph4.figure.set_size_inches(18.5, 10.5)
sns.despine()
```

```python
graph5 = sns.regplot(HDI_v, np.log(deaths2))
graph5.figure.set_size_inches(18.5, 10.5)
sns.despine()
```
